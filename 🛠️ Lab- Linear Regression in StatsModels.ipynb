## Step 2: Run a Simple Linear Regression with `TV` as the Predictor

As the analysis above indicates, `TV` looks like it has the strongest relationship with `sales`. Let's attempt to quantify that using linear regression.

```python
import statsmodels.api as sm
import pandas as pd

# Load the dataset
data = pd.read_csv('advertising.csv')

# Determine X and y values
X = data['TV']
y = data['sales']

# Add a constant to the predictor variable (required for statsmodels)
X = sm.add_constant(X)

# Create an OLS model
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
```

```python
import statsmodels.api as sm
import pandas as pd

# Load the dataset
data = pd.read_csv('advertising.csv')

# Determine X and y values
X = data['TV']
y = data['sales']

# Add a constant to the predictor variable (required for statsmodels)
X = sm.add_constant(X)

# Create an OLS model
model = sm.OLS(y, X).fit()

# Get model results and display the summary
results_summary = model.summary()
print(results_summary)
```

## Step 3: Evaluate and Interpret Results from Step 2

How does this model perform overall? What do the coefficients say about the relationship between the variables?

# Your answer here
Overall Model Performance
R-squared (0.612):

The R-squared value of 0.612 indicates that approximately 61.2% of the variance in sales can be explained by the TV advertising expenditure. This suggests a relatively strong linear relationship between TV advertising and sales.
Adjusted R-squared (0.610):

The adjusted R-squared value of 0.610 is very close to the R-squared value, indicating that the model is a good fit and the predictor (TV advertising) is significant.
F-statistic (312.1) and Prob (F-statistic) (1.47e-42):

The F-statistic tests the overall significance of the model. A high F-statistic value and a very low p-value (close to zero) indicate that the model is statistically significant.

## Step 4: Visualize Model with `TV` as Predictor

Create at least one visualization that shows the prediction line against a scatter plot of `TV` vs. sales, as well as at least one visualization that shows the residuals.

```python
# Plot the model fit (scatter plot and regression line)
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pandas as pd

# Load the dataset
data = pd.read_csv('advertising.csv')

# Determine X and y values
X = data['TV']
y = data['sales']

# Add a constant to the predictor variable (required for statsmodels)
X = sm.add_constant(X)

# Create an OLS model
model = sm.OLS(y, X).fit()

# Get predictions
predictions = model.predict(X)

# Plot the model fit (scatter plot and regression line)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['TV'], y=data['sales'], label='Actual Data')
sns.lineplot(x=data['TV'], y=predictions, color='red', label='Fitted Line')
plt.title('TV Advertising vs Sales with Regression Line')
plt.xlabel('TV Advertising Spend (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.show()

# Calculate residuals
residuals = y - predictions
```

```python
# Plot the model residuals
# Plot the model residuals
plt.figure(figsize=(10, 6))
sns.scatterplot(x=predictions, y=residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals of the Model')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('advertising.csv')

# Determine X and y values for radio
X_radio = data['radio']
y = data['sales']

# Add a constant to the predictor variable (required for statsmodels)
X_radio = sm.add_constant(X_radio)

# Create an OLS model for radio
model_radio = sm.OLS(y, X_radio).fit()

# Display results summary
results_summary_radio = model_radio.summary()
print(results_summary_radio)

# Get predictions for radio model
predictions_radio = model_radio.predict(X_radio)

# Plot the model fit (scatter plot and regression line) for radio
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['radio'], y=data['sales'], label='Actual Data')
sns.lineplot(x=data['radio'], y=predictions_radio, color='red', label='Fitted Line')
plt.title('Radio Advertising vs Sales with Regression Line')
plt.xlabel('Radio Advertising Spend (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.show()

# Calculate residuals for radio model
residuals_radio = y - predictions_radio

# Plot the model residuals for radio
plt.figure(figsize=(10, 6))
sns.scatterplot(x=predictions_radio, y=residuals_radio)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals of the Radio Model')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

## Step 6: Repeat Steps 2-4 with `newspaper` as Predictor

Once again, use this information to compare and contrast.

```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('advertising.csv')

# Determine X and y values for newspaper
X_newspaper = data['newspaper']
y = data['sales']

# Add a constant to the predictor variable (required for statsmodels)
X_newspaper = sm.add_constant(X_newspaper)

# Create an OLS model for newspaper
model_newspaper = sm.OLS(y, X_newspaper).fit()

# Display results summary
results_summary_newspaper = model_newspaper.summary()
print(results_summary_newspaper)

# Get predictions for newspaper model
predictions_newspaper = model_newspaper.predict(X_newspaper)

# Plot the model fit (scatter plot and regression line) for newspaper
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['newspaper'], y=data['sales'], label='Actual Data')
sns.lineplot(x=data['newspaper'], y=predictions_newspaper, color='red', label='Fitted Line')
plt.title('Newspaper Advertising vs Sales with Regression Line')
plt.xlabel('Newspaper Advertising Spend (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.show()

# Calculate residuals for newspaper model
residuals_newspaper = y - predictions_newspaper

# Plot the model residuals for newspaper
plt.figure(figsize=(10, 6))
sns.scatterplot(x=predictions_newspaper, y=residuals_newspaper)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals of the Newspaper Model')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

# Your interpretation here
The R-Squared here is the smallest we have seen yet: 0.05. This means that the model explains about 5% of the variance in sales. 5% is well below both the radio model (33%) and the TV model (61%).

The coefficient is also small, though similar to the TV coefficient. An increase of $1k in newspaper spending is associated with about 50 additional widget sales (more precisely, about 54.7). This is still much less than the 200-widget increase associated with $1k of additional radio spending.

<details>
    <summary style="cursor: pointer"><b>Answer (click to reveal)</b></summary>

Technically our model and coefficients are **still statistically significant** at an alpha of 0.05, but the p-values are much higher. For both the F-statistic (overall model significance) and the `newspaper` coefficient, our p-values are about 0.001, meaning that there is about a 0.1% chance that a variable with _no linear relationship_ would produce these statistics. That is a pretty small false positive rate, so we'll consider the model to be statistically significant and move on to interpreting the other results.

The R-Squared here is the smallest we have seen yet: 0.05. This means that **the model explains about 5% of the variance in `sales`**. 5% is well below both the `radio` model (33%) and the `TV` model (61%).

The coefficient is also small, though similar to the `TV` coefficient. **An increase of &dollar;1k in newspaper spending is associated with about 50 additional widget sales** (more precisely, about 54.7). This is still much less than the 200-widget increase associated with &dollar;1k of additional `radio` spending.

Visualizing this model, the best-fit line is clearly not a strong predictor. On the other hand, the residuals exhibit _homoscedasticity_, meaning that the distribution of the residuals doesn't vary much based on the value of `newspaper`. This contrasts with the `radio` and `TV` residuals which exhibit a "cone" shape, where the errors are larger as the x-axis increases. Homoscedasticity of residuals is a good thing, which we will describe more in depth when we discuss regression assumptions.
